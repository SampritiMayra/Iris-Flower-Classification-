 The following notebook uses **Support Vector Machines** on the famous Iris dataset.

```python
import matplotlib.pyplot as plt
import pandas as pd 
#%matplotlib inline 
import numpy as np
import seaborn as sns
from sklearn import datasets
import seaborn as sns
```

```python
#Importing Dataset
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names = ["sepal length", "sepal width", "petal length", "petal width", "species"])
iris.head()
```

```python
#Encoding the categorical column
iris = iris.replace({"species":  {"Iris-setosa":1,"Iris-versicolor":2, "Iris-virginica":3}})
iris.head()
```

```python
Y = iris['species'].values
X = iris.drop('species', axis = 1)
X.shape, Y.shape
```

```python
# Creating a pairplot to visualize the similarities and especially difference between the species
sns.pairplot(data=iris, hue='species')
```

```python
sns.heatmap(iris.corr(), vmin = -1, vmax = 1)
plt.title('Correlation On iris Classes', )
```

```python
from sklearn.model_selection import train_test_split #to split the dataset for training and testing

train_x, test_x, train_y, test_y = train_test_split(X,Y,test_size=0.2,random_state=1)

train_x.shape, train_y.shape
```

```python
from sklearn.svm import SVC
model=SVC(kernel = 'linear', random_state = 0)
```

```python
model.fit(train_x, train_y)
```

```python
y_pred = model.predict(test_x)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
#cm = confusion_matrix(test_y, y_pred)
#print(cm)

from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = model, X = train_x, y = train_y, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
model=SVC(kernel = 'rbf', random_state = 0)
model.fit(train_x, train_y)
```

```python
y_pred = model.predict(test_x)
y_pred
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = model, X = train_x, y = train_y, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
model=SVC(kernel = 'poly', random_state = 0)
model.fit(train_x, train_y)
```

```python
y_pred = model.predict(test_x)
y_pred
```
